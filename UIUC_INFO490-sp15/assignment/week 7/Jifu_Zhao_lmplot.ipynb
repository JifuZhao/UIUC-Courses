```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
#!head -10 '/data/airline/2001.csv'
```

```python
def get_columns(filename):
    '''
    this function is going to select the columns named 'AirTime','Distance' and 'DayOfWeek'
    '''
    
    df = pd.read_csv(filename, encoding='latin1',usecols = ['AirTime','Distance', 'DayOfWeek']).dropna(axis=0, how = 'any')

    return df
```

```python
# This may take some time

filename = '/data/airline/2001.csv'
df = get_columns(filename)
```

```python
# This process may take some time
# to save time, I will not plot all datas

mpl.rcParams['figure.figsize']=[12,5]
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
sns.regplot('Distance', 'AirTime', df[::1000], ax=ax1)
sns.boxplot(df['AirTime'][::10000], df['DayOfWeek'][::10000], color="Blues_r", ax=ax2)

ax1.set_xlabel('Distance')
ax1.set_xlim(0, 4500)
ax1.set_ylabel('Airtime')
ax1.set_ylim(0, 500)
ax1.set_title('The scatter plot of Airtime vs. Distance \n by Jifu Zhao')

ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_xticklabels(['Monday', 'Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
ax2.set_title('The box plot of daily Airtime \n by Jifu Zhao')
fig.tight_layout()
plt.show()
```
