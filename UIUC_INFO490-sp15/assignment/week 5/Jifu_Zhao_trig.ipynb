```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

x = np.linspace(-np.pi, np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

ax.plot(x, y1, color = 'red', label = 'sin(x)')
ax.plot(x, y2, color = 'blue', label = 'cos(x)')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(-5, 5)
ax.set_ylim(-1.1, 1.1)
ax.set_title("Plot of y = sin(x) and y = cos(x)\n by Jifu Zhao")

plt.legend()
plt.show()
```
