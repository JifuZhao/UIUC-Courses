```python
import warnings
warnings.simplefilter('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from k_means import kMeans
from k_centers import kCenters
from single_swap import singleSwap
from spectral_clustering import spectralClustering
from EM import EM

%matplotlib inline
```

```python
# read the data
clustering = pd.read_csv('./data/clustering.csv', header=None).values
bigClustering = pd.read_csv('./data/bigClusteringData.csv', header=None).values

# define colors
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', '#7FFFD4', '#9ACD32', '#FFA500']
```

# I. K-Means Algorithm

```python
# clustering.txt
X = clustering
K = 3
tol = 10e-5
Y, C, D = kMeans(X, K, tol=tol, random_state=0, verbose=True)
print('Cost is:\t', D)

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[C==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            alpha=0.6, markersize=3, label='Cluster ' + str(i+1))
    ax.plot(Y[i, 0], Y[i, 1], '*', color='k', markersize=10)
ax.set_title('K-means for clustering.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
fig.savefig('./result/pdf/clustering_kMeans_' + str(K) + '.pdf')
fig.savefig('./result/clustering_kMeans_' + str(K) + '.png', dpi=300)
plt.show()

# bigClusteringData.txt
X = bigClustering
K = 3
tol = 10e-5
Y, C, D = kMeans(X, K, tol=tol, random_state=0, verbose=True)
print('Cost is:\t', D)

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[C==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            alpha=0.6, markersize=3, label='Cluster ' + str(i+1))
    ax.plot(Y[i, 0], Y[i, 1], '*', color='k', markersize=10)
ax.set_title('K-means for bigClusteringData.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
fig.savefig('./result/pdf/bigClustering_kMeans_' + str(K) + '.pdf')
fig.savefig('./result/bigClustering_kMeans_' + str(K) + '.png', dpi=300)
plt.show()
```

# V. EM Algorithm

```python
# clustering.txt
X = clustering
K = 3
threshold = 1e-7
em = EM(m=K, threshold=threshold, random_state=0, maxIter=500)
em.train(X)  # train the EM model

# get the label
label = em.get_label()  # label for each x
mu = em.mu  # cluster centers
D = em.D
print('Cost is:\t', D)

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[label==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            alpha=0.6, markersize=4, label='Cluster ' + str(i+1))
    ax.plot(mu[i][0], mu[i][1], '*', color='k', markersize=10)
ax.set_title('EM for clustering.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
fig.savefig('./result/pdf/clustering_EM_' + str(K) + '.pdf')
fig.savefig('./result/clustering_EM_' + str(K) + '.png', dpi=300)
plt.show()

# bigClusteringData.txt
X = bigClustering
K = 3
threshold = 1e-7
em = EM(m=K, threshold=threshold, random_state=0, maxIter=500)
em.train(X)  # train the EM model

# get the label
label = em.get_label()  # label for each x
mu = em.mu  # cluster centers
D = em.D
print('Cost is:\t', D)

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[label==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            alpha=0.5, markersize=3, label='Cluster ' + str(i+1))
    ax.plot(mu[i][0], mu[i][1], '*', color='k', markersize=10)
ax.set_title('EM for bigClusteringData.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
fig.savefig('./result/pdf/bigClustering_EM_' + str(K) + '.pdf')
fig.savefig('./result/bigClustering_EM_' + str(K) + '.png', dpi=300)
plt.show()
```
