```python
import warnings
warnings.simplefilter('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from k_means import kMeans
from k_centers import kCenters
from single_swap import singleSwap
from spectral_clustering import spectralClustering
from EM import EM

%matplotlib inline
```

```python
# read the data
clustering = pd.read_csv('./data/clustering.csv', header=None).values
bigClustering = pd.read_csv('./data/bigClusteringData.csv', header=None).values

# define colors
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', '#7FFFD4', '#9ACD32', '#FFA500']
```

# I. K-Means Algorithm

```python
# clustering.txt
X = clustering
K = 3
tol = 10e-5
Y, C, D = kMeans(X, K, tol=tol, random_state=0, verbose=True)

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[C==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            markersize=4, label='Cluster ' + str(i+1))
ax.set_title('K-means for clustering.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()

# bigClusteringData.txt
X = bigClustering
K = 3
tol = 10e-5
Y, C, D = kMeans(X, K, tol=tol, random_state=0, verbose=True)

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[C==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            markersize=4, label='Cluster ' + str(i+1))
ax.set_title('K-means for bigClusteringData.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()
```

# II. Greedy K-Centers Algorithm

```python
# clustering.txt
X = clustering
K = 3
Q, D = kCenters(X, K, random_state=0, verbose=True)

# plot the clustered data
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.', color='g', markersize=4)
ax.plot(Q[:, 0], Q[:, 1], 'o', color='r', markersize=8)
ax.set_title('K-centers for clustering.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()

# bigClusteringData.txt
X = bigClustering
K = 3
Q, D = kCenters(X, K, random_state=0, verbose=True)

# plot the clustered data
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.', color='g', markersize=4)
ax.plot(Q[:, 0], Q[:, 1], 'o', color='r', markersize=8)
ax.set_title('K-centers for bigClusteringData.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()
```

# III. Single-Swap Algorithm

```python
# clustering.txt
X = clustering
K = 3
Q, D = singleSwap(X, K, tau=0.05, random_state=0, verbose=True)

# plot the clustered data
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.', color='g', markersize=4)
ax.plot(Q[:, 0], Q[:, 1], 'o', color='r', markersize=8)
ax.set_title('Single-Swap for clustering.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()

# bigClusteringData.txt
X = bigClustering
K = 3
Q, D = singleSwap(X, K, tau=0.05, random_state=0, verbose=True)

# plot the clustered data
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.', color='g', markersize=4)
ax.plot(Q[:, 0], Q[:, 1], 'o', color='r', markersize=8)
ax.set_title('Single-Swap for bigClusteringData.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()
```

# V. EM Algorithm

```python
# clustering.txt
X = clustering
K = 3
threshold = 1e-7
em = EM(m=K, threshold=threshold, random_state=0, maxIter=500)
em.train(X)  # train the EM model

# get the label
label = em.get_label()

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[label==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            markersize=4, label='Cluster ' + str(i+1))
ax.set_title('EM for clustering.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()

# bigClusteringData.txt
X = bigClustering
K = 3
threshold = 1e-7
em = EM(m=K, threshold=threshold, random_state=0, maxIter=500)
em.train(X)  # train the EM model

# get the label
label = em.get_label()

# plot the clustered data
fig, ax = plt.subplots()
for i in range(K):
    tmp = X[label==i, :]
    ax.plot(tmp[:, 0], tmp[:, 1], '.', color=colors[i], 
            markersize=4, label='Cluster ' + str(i+1))
ax.set_title('EM for bigClusteringData.txt with K=' + str(K))
ax.legend(fontsize=8, loc='best')
plt.show()
```
