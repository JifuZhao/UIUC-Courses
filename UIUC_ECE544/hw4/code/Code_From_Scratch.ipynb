```python
import warnings
warnings.simplefilter('ignore')

import numpy as np
import matplotlib.pyplot as plt
from mnist import MNIST

from RBM import RBM

%matplotlib inline
```

```python
# load MNIST dataset
mndata = MNIST('./data/')

train_img, train_label = mndata.load_training()
train_img = np.array(train_img)

test_img, test_label = mndata.load_testing()
test_img = np.array(test_img)

# transform input data into binary values
data = (train_img > 0).astype(float)
test_data = (test_img > 0).astype(float)
```

```python
# build the Restricted Boltzmann Machine
model = RBM(hidden_nodes=200, learning_rate=0.1, batch_size=10, n_iter=5, verbose=1)

model.train(data, train_mse=True, test_data=test_data, frequency=10000)
W, b, c = model.get_params()

# visualize the learned weights
imgs = np.reshape(W[:200, :], (200, 28, 28))

fig, ax = plt.subplots(nrows=8, ncols=8)
for i in range(8):
    for j in range(8):
        ax[i, j].imshow(imgs[i * 8 + j, :, :], cmap=plt.cm.gray_r, interpolation='none')
        ax[i, j].set_xticks([])
        ax[i, j].set_yticks([])
        ax[i, j].axis('image')
plt.tight_layout(pad=0, h_pad=0, w_pad=0, rect=None)
plt.show()
```

# Test case for single case

```python
# build the Restricted Boltzmann Machine
model = RBM(hidden_nodes=200, learning_rate=0.1, batch_size=1, n_iter=10, verbose=1)

model.train(data, train_mse=True, test_data=test_data, frequency=10000)
W, b, c = model.get_params()

# visualize the learned weights
imgs = np.reshape(W[:200, :], (200, 28, 28))

fig, ax = plt.subplots(nrows=8, ncols=8)
for i in range(8):
    for j in range(8):
        ax[i, j].imshow(imgs[i * 8 + j, :, :], cmap=plt.cm.gray_r, interpolation='none')
        ax[i, j].set_xticks([])
        ax[i, j].set_yticks([])
        ax[i, j].axis('image')
plt.tight_layout(pad=0, h_pad=0, w_pad=0, rect=None)
plt.show()

# plot the training and test error
fig, ax = plt.subplots()
ax.plot(model.train_error, 'g.-', label='Training Error')
ax.plot(model.test_error, 'r.-', label='Testing Error')
ax.legend(fontsize=10)
plt.show()
```
