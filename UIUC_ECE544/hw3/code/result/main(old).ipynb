```python
import warnings
warnings.simplefilter('ignore')

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import copy

from EM import EM

%matplotlib inline
```

```python
# read the image
img = plt.imread('./corgi.png')[:, :, :3]

# reshape the input vector
x = np.reshape(img, (-1, 3))
x = x / np.max(x)
```

```python
np.random.seed(12)  # 7, 10, 12, 30, 70
# train the EM model
m = 3
em = EM(m=m, threshold=0.0001, maxIter=100)
em.train(x)

# get all parameters
w, mu, sigma, logLikelihood = em.get_params()
# get the label
label = em.get_label()

# # plot the likelihood
# plt.figure(figsize=(4, 3))
# plt.plot(logLikelihood, 'o-', markersize=3)
# plt.show()

# # plot the clustered data
# fig = plt.figure(figsize=(6, 5))
# ax = fig.add_subplot(111, projection='3d')
# for i in range(1, m + 1):
#     data = x[label == i]
#     ax.plot(data[:, 0], data[:, 1], data[:, 2], '.', markersize=0.1, label='Class ' + str(i))
# ax.set_xlabel('Red')
# ax.set_ylabel('Green')
# ax.set_zlabel('Blue')
# plt.show()

# reconstruct the image
test = copy.deepcopy(x)
for i in range(1, 4):
    test[label == i] = mu[i - 1]
    
fig = plt.figure()
plt.imshow(np.reshape(test, img.shape))
plt.title('Reconstruction with m = 3')
plt.xticks([])
plt.yticks([])
fig.savefig('./result/corgi_3.png', dpi=300)
plt.show()

print('w:\n', np.round(w, 5))
print('\nmu:\n', np.round(mu, 5))
print('\nsigma:\n', np.round(sigma, 5))
```
